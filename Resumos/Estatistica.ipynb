```python
print(np.mean(flowers['sepal_length']))
print(np.mean(flowers['sepal_width']))
print(np.mean(flowers['petal_length']))
print(np.mean(flowers['petal_width']))
```

#### Mediana

```python
print(np.median(flowers['sepal_length']))
print(np.median(flowers['sepal_width']))
print(np.median(flowers['petal_length']))
print(np.median(flowers['petal_width']))
```

#### Moda

```python
print((flowers['sepal_length']).mode())
print((flowers['sepal_width']).mode())
print((flowers['petal_length']).mode())
print((flowers['petal_width']).mode())
```

#### Separatrizes

```python
print(np.quantile(flowers['sepal_length'], 0.1)) #Percentil 10
print(np.quantile(flowers['sepal_length'], 0.5)) #Percentil 50 - Igual a mediana
print(np.quantile(flowers['sepal_length'], 0.25)) #Primeiro Quartil - 25%
```

### Aplicações em Data Science
#### Método de Monte Carlo

```python
T = 99999999
I = 0
for i in range(T):
    x = random.random()
    y = random.random()
    if(x**2 + y**2) < 1:
        I = I + 1
```

```python
print(4 * I / T)
```
